## **🔻 Correlation Matrix & Heatmap**

To evaluate the relationships between different numerical variables, we compute a **correlation matrix**. This matrix gives insights into the strength and direction of relationships among variables, identifying:

- **Strong positive correlation**: Values close to **+1**

   - **Insight**: The number of bathrooms also has a significant impact on the price. More bathrooms typically indicate a higher price, possibly due to the increased amenities and comfort.

3. **`bedrooms` vs. `price`**:
   
   - **Correlation**: **0.37** (Moderate positive correlation)
   - **Insight**: There is a moderate relationship between the number of bedrooms and the price. While more bedrooms can increase the value, the effect is less pronounced than with the area or number of bathrooms.

4. **`stories` vs. `price`**:
   
   - **Correlation**: **0.42** (Weak to moderate positive correlation)
   - **Insight**: The number of stories has a relatively weaker impact on the price. This may indicate that while multi-story homes can add value, it is not as strong a driver compared to size and amenities.

5. **`parking` vs. `price`**:
   
   - **Correlation**: **0.38** (Weak positive correlation)
   - **Insight**: The availability of parking spaces shows a weak relationship with the price. It suggests that while parking can be a desirable feature, it does not heavily influence the pricing in comparison to other factors like area or bathrooms.

- **Strongest driver** of house prices: **Area (0.54)**

## Handling Outliers

###**Box Plot for Numeric Columns**

In this step, we create box plots for all numeric columns in the dataset to visually identify potential outliers. The box plot provides a graphical representation of the distribution of the data, highlighting the median, quartiles, and any data points that fall outside the expected range.

```python
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#ff99cc', '#99ccff']

for i, color in zip(num.columns, colors):
    plt.figure(figsize=(6, 4))
    plt.boxplot(num[i], vert=False, patch_artist=True, boxprops=dict(facecolor=color, color=color))
    plt.title(f'Boxplot of {i}')
    plt.xlabel(i)
    plt.show()
```

# Encoding



❐ **Objective**: Convert categorical columns into a suitable numerical format to enhance model performance.

❐ **Purpose**: Improve the performance of the data analysis model by representing categorical values as numbers.

### Encoding Type Used


**Label Encoding**  

    We will use Label Encoding to transform categorical values into numerical representations
    facilitating better understanding by machine learning models.

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()

for column in cat.columns:
  data[column] = LE.fit_transform(data[column])
```

# Scaling

We will use **MinMaxScaler** to scale all the data, as the values in the dataset vary significantly across different columns. While some columns, such as those containing property prices and areas, have large values, the other columns, which were encoded, also benefit from scaling to ensure consistency across the dataset.

The goal of this process is to improve the accuracy of statistical models and machine learning by transforming the data into a specific range, which enhances the models' performance and makes them more reliable.

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
data_normalized = pd.DataFrame(scaler.fit_transform(data), columns=data.columns)
```
